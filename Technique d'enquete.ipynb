```python
from statsmodels.tools import add_constant
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegressi
```

```python
df["Salaire"] = df["SALPRS_15"].apply(lambda x:  1 if x > 0 else 0)
```

```python
df["Q1"] = df["Q1"] - 1
```

```python
df["ENFANT"] = df["CA20_15"] - 1
```

```python
X = df[["AGE10","Q1", "ENFANT" ]]
y = df["Salaire"]
XRegBis = sm.tools.add_constant(X)
Model = sm.Logit(y, XRegBis.astype(float))
res = Model.fit()
print(res.summary2())
```

```python
region_groupe = pd.get_dummies(df['CA0AREGIONF_15']).drop(['**', 'ALSACE'], axis=1)
contrat_groupe = pd.get_dummies(df['CONTRAT_15']).drop('05 Intérim', axis=1)
```

```python
region_groupe['outre mer'] = region_groupe.apply(lambda z: z['MAYOTTE']+z['LA RÉUNION']+z['GUYANE']+z['MARTINIQUE']+z['GUADELOUPE'], axis=1)
region_groupe = region_groupe.drop(['MAYOTTE', 'LA RÉUNION', 'GUYANE', 'MARTINIQUE', 'GUADELOUPE'], axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df[['AGE10']] = sc.fit_transform(df[['AGE10']])
```

```python
x_var = pd.concat([df.AGE10, df.Q1, region_groupe, contrat_groupe], axis=1)
x_var.head()
```

```python
salaire_mensuel = df['SALPRS_15']
mask = ~salaire_mensuel.isnull()
```

```python
salaire_mensuel
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(x_var[mask], salaire_mensuel[mask])

salaires_estimés = reg.predict(x_var[~mask])
```
